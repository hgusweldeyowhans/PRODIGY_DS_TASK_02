# droping unnecessary columns 

```python
data.drop(columns = "Cabin" ,axis = 1,inplace = True)
```

# Filling the missing values of the age of column by mean values of column

```python
data['Age'].fillna(data['Age'].mean() ,inplace = True)
```

# filling the missing value of the embarked column by mode value of the column

```python
data['Embarked'].fillna(data['Embarked'].mode()[0] ,inplace = True)
```

# visualization of survival rate in different age category

```python
cut_points = [0,5,12,18,35,60,100]
labels_names = ['Infant','Child','Teenager','Young Adult','Adult','Senior']
data['Age_categories'] = pd.cut(data['Age'] ,bins = cut_points , labels = labels_names ,right  = False)
age_cat_pivot = data.pivot_table(index = 'Age_categories',values ='Survived')
colors = ['blue','green','orange','purple','pink','brown']
fig ,ax = plt.subplots()
bars = ax.bar(age_cat_pivot.index,age_cat_pivot['Survived'] , color = colors)
handles = [plt.Rectangle((0,0),1,1, color =colors[i]) for i in range(len(colors))]
ax.legend(handles ,labels_names)
ax.set_title('Survival Rates by Age Category')
ax.set_xlabel('Age Category')
ax.set_ylabel('Survival Rate')
plt.show()
```
